```python
from matplotlib import pyplot as pt

f = open('plots.txt')
cnt = f.readlines()
f.close()

tx_rate = []
trx1 = []
trx2 = []
rx_rate1 = []
rx_rate2 = []

for line in cnt:
    if '-----' in line:
        break
    ls = line.split()
    tx_rate.append(float(ls[0]))
    trx1.append(float(ls[1]))
    rx_rate1.append(float(ls[2]))
    trx2.append(float(ls[3]))
    rx_rate2.append(float(ls[4]))
```

```python
pt.rcParams["figure.figsize"] = (20,10)
pt.plot(tx_rate,trx1,'-Dr',label='1 Rx - 1 Tx',markersize=25)
pt.plot(tx_rate,trx2,'-ob',label='1 Rx - 2 Tx',markersize=25)
pt.xticks(fontsize=30)
pt.yticks(fontsize=30)
pt.xlabel('Tx rate [msg/s]',fontsize=30)
pt.ylabel('$T_{S}$ [ms]',fontsize=30)
pt.grid(True)
pt.legend(fontsize=30)
pt.savefig('trx_vs_tx_rate.pdf',format='pdf',dpi=300, bbox_inches='tight')
pt.show()
```

```python
pt.rcParams["figure.figsize"] = (20,10)
pt.plot(tx_rate,rx_rate1,'-Dr',label='1 Rx - 1 Tx',markersize=25)
pt.plot(tx_rate,rx_rate2,'-ob',label='1 Rx - 2 Tx',markersize=25)
pt.xticks(fontsize=30)
pt.yticks(fontsize=30)
pt.xlabel('Tx rate [msg/s]',fontsize=30)
pt.ylabel('Rx rate [msg/s]',fontsize=30)
pt.grid(True)
pt.legend(fontsize=30)
pt.savefig('rx_rate_vs_tx_rate.pdf',format='pdf',dpi=300, bbox_inches='tight')
pt.show()
```
