```python
from matplotlib import pyplot as pt
```

```python
P = {}
for i in range(1,11):
    P[i]=[1-i*.05, 0, i*.05]

col_theor = [[0 for j in range(100)] for i in range(10)]
for i in range(1,11):
    for j in range(1,100):    
        col_theor[i-1][j-1]=1-(1-P[i][2]/15)**(j)
    
    
    
```

```python
print(P)
```

```python
pt.rcParams["figure.figsize"] = (10,10)
pt.plot(col_theor[0])
pt.plot(col_theor[1])
pt.plot(col_theor[2])
pt.plot(col_theor[3])
pt.plot(col_theor[4])
pt.plot(col_theor[5])
pt.plot(col_theor[6])
pt.plot(col_theor[7])
pt.plot(col_theor[8])
pt.plot(col_theor[9])
pt.axis([0,100,0,1.0])
pt.xticks(range(0,100,5), [str(n) for n in range(0,100,5)])
pt.show()
```
