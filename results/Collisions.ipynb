## From Simulation

```python
evts = {}

for i in range(0,20):
    f1 = open("r4-d%d-result.txt"%i)
    raw1 = f1.read()
    f1.close()
    
    for e in range(len(raw1)):
        if raw1[e] == 'B':
            try:
                evts[e].append(i)
            except:
                evts[e] = [i]
    print('Done with ',i)
succ_tx = []
for k in evts:
    if len(evts[k]) == 1:
        succ_tx.append(k)

print(len(succ_tx)/len(evts))


```
