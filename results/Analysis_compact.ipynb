```python
from analysis import *
from matplotlib import pyplot as pt
import subprocess
import cProfile
```

```python

subprocess.run('tutorial.exe 9')
hists = {}

for i in range(1,9):
    f = open('r'+str(i)+'-analysis.txt')
    r1 = f.readlines()
    f.close()
    evts_raw = r1[0].split(';')
    Eb = int(evts_raw[0].split(':')[1])
    succ =int(r1[1].split(':')[1])
    print("P(suc)=",succ/Eb)
    hists[i] = []
    for j in range(2,30):
        try:
            rs = r1[j].split(':')
            hists[i].append(float(rs[1])/1000)
        except:
            break
```

```python
pt.rcParams["figure.figsize"] = (20,10)
pt.plot(range(len(hists[1])),hists[1])
pt.plot(range(len(hists[2])),hists[2])
pt.plot(range(len(hists[3])),hists[3])
pt.plot(range(len(hists[4])),hists[4])
pt.plot(range(len(hists[5])),hists[5])
pt.plot(range(len(hists[6])),hists[6])
pt.plot(range(len(hists[7])),hists[7])
pt.plot(range(len(hists[8])),hists[8])
#pt.legend()
#pt.axis([0,30,0,0.3])
pt.xticks(range(30), [str(n) for n in range(30)])
pt.xlabel(r'$\mathcal{k}$', fontsize = 18)
pt.ylabel(r'P(k-messages-received)')
pt.savefig('test.svg',format='svg',dpi=300, bbox_inches='tight')
pt.grid(True)
pt.show()
```
